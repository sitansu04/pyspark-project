```python
import pandas as pd
from pyspark.sql import SparkSession
from pyspark import *
```

```python
spark = SparkSession.builder.getOrCreate()
```

```python
df = spark.read.option("header",True).option("inferSchema",True).csv("/content/pyspark-project/ipl_2022_dataset.csv")
df.show()
```

```python
df.printSchema()
```

```python
print("rows:- ",df.count())
print("columns:-", len(df.columns))
```

```python
 print(len(df.select("Base Price").distinct().collect()))
```

```python
df.groupBy("Base Price").count().show()
```

```python
df2 = df.withColumnRenamed("Player","player").withColumnRenamed("Base Price","base_price").withColumnRenamed("TYPE","type").withColumnRenamed("COST IN ₹ (CR.)","cost_inr").withColumnRenamed("Cost IN $ (000)","cost_usd").withColumnRenamed("2021 Squad","2021_team").withColumnRenamed("Team","2022_Team")
```

```python
df2.show()
```

```python
df3 = df2.drop("cost_usd")
```

```python
df3.show()
```

```python
df3.select("player","cost_inr").filter(df3.type=="BATTER").orderBy("cost_inr",ascending=False).show(3)
```

```python
df3.createOrReplaceTempView("ipl")
```

```python
spark.sql("SELECT player, cost_inr FROM ipl WHERE type='BATTER' ORDER BY cost_inr desc LIMIT 3").show()
```

```python
spark.sql("select player, cost_inr from ipl where type='BOWLER' order by cost_inr desc limit 5").show()
```

```python
df3.select("player","cost_inr").filter(df3.type=="BOWLER").orderBy("cost_inr",ascending=False).show(5)
```

```python
df3.filter(df3.type=="ALL-ROUNDER").orderBy(df3.cost_inr,ascending=False).show(5)
```

```python
df3.select("player","cost_inr").filter(df3.type=="WICKETKEEPER").orderBy(df3.cost_inr,ascending=True).show(5)
```

```python
spark.sql("select player, cost_inr from ipl where type=='WICKETKEEPER' and cost_inr is not null order by cost_inr asc").show(5)
```

```python
spark.sql("select type,round(avg(cost_inr),2) as avg_cost from ipl group by type").show()
```

```python
from pyspark.sql import functions as f
df3.groupBy("type").agg(f.round(f.mean("cost_inr"),2).alias("avg_cost")).show()
```

```python
spark.sql("select player,2021_team,cost_inr as salary from ipl where base_price=='Retained' order by salary desc").show(5)
```

```python
df3.select("player","2021_team","cost_inr").filter(df3.base_price=="Retained").orderBy("cost_inr",ascending=False).show(5)
```
