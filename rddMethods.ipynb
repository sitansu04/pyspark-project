```python
from pyspark import SparkContext
sc = SparkContext.getOrCreate()
rdd = sc.parallelize([1, 2, 3, 4, 5])
mapped_rdd = rdd.map(lambda x: x * 2)
print(mapped_rdd.collect())
```

```python
filtered_rdd = rdd.filter(lambda x: x % 2 == 0)
print(filtered_rdd.collect())
```

```python
from functools import reduce
result = rdd.reduce(lambda x, y: x + y)
print(result)
```

```python
distinct_rdd = rdd.distinct()
print(distinct_rdd.collect())
```

```python
pair_rdd = rdd.map(lambda x: (x % 2, x))
grouped_rdd = pair_rdd.groupByKey()
print([(k, list(v)) for k, v in grouped_rdd.collect()])
print(grouped_rdd.collect())
```

```python
reduced_rdd = pair_rdd.reduceByKey(lambda x, y: x + y)
print(reduced_rdd.collect())
```

```python
sorted_rdd = pair_rdd.sortByKey(ascending=False)
print(sorted_rdd.collect())
```

```python
flat_mapped_rdd = rdd.flatMap(lambda x: (x, x * 2))
print(flat_mapped_rdd.collect())
```

```python
rdd1 = sc.parallelize([1, 2, 3])
rdd2 = sc.parallelize([3, 4, 5])
unioned_rdd = rdd1.union(rdd2)
print(unioned_rdd.collect())
```

```python
intersected_rdd = rdd1.intersection(rdd2)
print(intersected_rdd.collect())
```

```python
subtracted_rdd = rdd2.subtract(rdd1)
print(subtracted_rdd.collect())
```

```python
rdd3 = sc.parallelize([(1, "A"), (2, "B"), (3, "C")])
rdd4 = sc.parallelize([(1,"d"),(2,"e"),(3,"f")])
sorted_rdd3 = rdd3.sortByKey()
joined_rdd = rdd3.join(rdd4)
# print(joined_rdd.collect())
print(sorted_rdd3.collect())
```

```python
coalesced_rdd = rdd.coalesce(4)
print(coalesced_rdd.getNumPartitions())
```

```python
repartitioned_rdd = rdd.repartition(6)
print(repartitioned_rdd.collect())
```
